# Train ,Test and split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
tr_acc=[]
ts_acc=[]
for k in range (1,15):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    train_accuracy=knn.score(X_train,y_train)
    test_accuracy=knn.score(X_test,y_test)
    tr_acc.append(train_accuracy)
    ts_acc.append(test_accuracy)
```

```python
plt.plot(tr_acc, label = 'Training accuracy')
plt.plot(ts_acc, label = 'Test accuracy')
plt.xlabel('Train Accuracy')
plt.ylabel('Test Accuracy')
plt.show()
```

```python
#we can see that k=4 produces the most accurate results
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
```

```python
knn.fit(X_train, y_train)
y_pred_KNeighborsClassifier = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
scores = []
cv_scores = []
score = accuracy_score(y_pred_KNeighborsClassifier,y_test)
scores.append(score)
```

```python
train_pred = knn.predict(X_train)
test_pred= knn.predict(X_test)
```

```python
# Confusion matrix of trained data
from sklearn.metrics import confusion_matrix,classification_report
confusion_matrix(y_train,train_pred)
```

```python
# Classification Report of trained data
print(classification_report(y_train,train_pred))
```
