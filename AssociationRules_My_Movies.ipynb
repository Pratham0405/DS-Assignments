```python
df = movie.iloc[:,5:]
df
```

# Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

# 1. Association rules with 10% Support and 70% confidence

```python
freq = apriori(df,min_support=0.1,use_colnames=True
)
freq
```

```python
# 70% confidence
ass_rules = association_rules(freq,metric='lift',min_threshold=0.7)
ass_rules
```

```python
ass_rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
ass_rules[ass_rules.lift>1]
```

```python
ass_rules[['support','confidence','lift']].hist()
```

```python
# visualization of obtained rule
plt.scatter(ass_rules['support'],ass_rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
# visualization of obtained rule
import seaborn as sns
sns.scatterplot(x = 'support', y = 'confidence', data=ass_rules, hue='antecedents')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

# 2.Association rules with 5% Support and 90% confidence

```python
# with 5% support
frequent_itemsets2=apriori(df,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

```python
# 90% confidence
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules2
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules2[rules2.lift>1]
```

```python
rules2[['support','confidence','lift']].hist()
```
